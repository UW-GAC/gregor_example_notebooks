Need to get the actual paths to LR bams for the cases we want. The aligned tables contain the cloud paths to the bam files. They also have a column called `experiment_pac_bio_id` or `experiment_nanopore_id` which is how we can join to the experiment table. Experiment table can be selected based on ??? the `ont_ids` and `pacbio_ids`.

```python
ont_aligned  = pd.read_csv('~/GREGoR_WatershedSV/aligned_nanopore.tsv', sep = "\t")
pacbio_aligned = pd.read_csv('~/GREGoR_WatershedSV/aligned_pac_bio.tsv', sep = "\t")
```

```python
ont_paths = pd.merge(ont_aligned, ont_experiment, left_on = "experiment_nanopore_id", right_on = "entity:experiment_nanopore_id", how = "inner")
ont_paths = ont_paths[ont_paths.participant_id.isin(ont_ids)]
ont_paths = ont_paths[["participant_id", "experiment_nanopore_id", "aligned_nanopore_file", "aligned_nanopore_index_file"]]
display(ont_paths)
print(ont_paths.shape)
#ont_paths.to_csv('~/GREGoR_WatershedSV/ont_paths.csv', index=False, na_rep='NA')
```

```python
pacbio_paths = pd.merge(pacbio_aligned, pacbio_experiment, left_on = "experiment_pac_bio_id", right_on = "entity:experiment_pac_bio_id", how = "inner")
pacbio_paths = pacbio_paths[pacbio_paths.participant_id.isin(pacbio_ids)]
pacbio_paths = pacbio_paths[["participant_id", "experiment_pac_bio_id", "aligned_pac_bio_file", "aligned_pac_bio_index_file"]]
display(pacbio_paths)
pacbio_paths.shape
pacbio_paths.to_csv('~/GREGoR_WatershedSV/pacbio_paths.csv', index=False, na_rep='NA')
```

Some cases where we have multiple sequencing options

```python
cases_to_decide_paths = pd.concat([ont_paths[ont_paths.participant_id.isin(cases_to_decide)],pacbio_paths[pacbio_paths.participant_id.isin(cases_to_decide)]]).sort_values("participant_id")
cases_to_decide_paths["bam_path"] = cases_to_decide_paths["aligned_nanopore_file"].fillna(cases_to_decide_paths["aligned_pac_bio_file"]) 
display(cases_to_decide_paths[["participant_id", "bam_path"]])
cases_to_decide_paths = cases_to_decide_paths[["participant_id", "bam_path"]]
cases_to_decide_paths.to_csv('~/GREGoR_WatershedSV/cases_to_decide_paths.csv', index=False, na_rep='NA')
```

It turns out that everything that's duplicated is just a duplicate entry not actually new sequencing. The only exception is for the three cases that have both ONT and pacbio sequencing. In these cases, we'll go with the ONT since the size of the bams is considerably larger which should translate to increased coverage.

```python
ont_choices = ["PMGRC-157-146-1", "PMGRC-168-167-2", "PMGRC-169-167-1"]
ont_paths = ont_paths.drop_duplicates()
display(ont_paths)
ont_paths.shape
```

```python
pacbio_choices = pacbio_paths[~pacbio_paths.participant_id.isin(ont_choices)]
pacbio_paths = pacbio_paths[~pacbio_paths.participant_id.isin(ont_choices)] #Table is only cases where we want the pacbio data
pacbio_paths = pacbio_paths.drop_duplicates()
display(pacbio_paths)
pacbio_paths.shape
```

```python
watershedsv_bamlist = pd.concat([ont_paths, pacbio_paths])
watershedsv_bamlist["bam_path"] = watershedsv_bamlist["aligned_nanopore_file"].fillna(watershedsv_bamlist["aligned_pac_bio_file"])
watershedsv_bamlist["index_path"] = watershedsv_bamlist["aligned_nanopore_index_file"].fillna(watershedsv_bamlist["aligned_pac_bio_index_file"])
watershedsv_bamlist = watershedsv_bamlist[["participant_id", "bam_path", "index_path"]]
watershedsv_bamlist #This is the final list of bams used for analysis
```

Convert the bamlist to a samplesheet for Tanner's SV pipeline

```python
oak_path = "/oak/stanford/groups/smontgom/pfp/GREGoR_WatershedSv/data/"

samplesheet = watershedsv_bamlist.copy(deep=True)
samplesheet["oak_path"] = oak_path + samplesheet["bam_path"].str.split("/").str[-1]
samplesheet = samplesheet.rename(columns = {"participant_id": "SampleName", "oak_path": "AlignedBam"})
samplesheet["SampleDir"] = pd.NA
samplesheet = samplesheet[["SampleName", "SampleDir", "AlignedBam"]]
samplesheet
```
